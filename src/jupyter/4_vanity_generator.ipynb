```python
from econia.account import Account
```

```python
vanity = 'b1d4c0de' # Initial hex characters
```

```python
path = f'../../.secrets/vanity/{vanity}.key'
length = len(vanity)
```

```python
%%time
account = Account()
while(account.address()[0:length] != vanity):
    account = Account()
```

```python
print(f'Account address:\n{account.address()}')
account.save_seed_to_disk(path)
assert Account(path=path).address() == account.address()
print(f'Saved to disk')
```
